# Away Analysis

How many times have the Rangers played against Florida? 

```python
games_df[
    (games_df["team"] == "NYR") &
    (games_df["season"] == 2023) &
    (games_df["opposingTeam"] == "FLA")
]["gameId"].nunique()
```

```python
unique_game_id = games_df[
    (games_df["team"] == "NYR") &
    (games_df["season"] == 2023) &
    (games_df["opposingTeam"] == "FLA") &
    (games_df["home_or_away"].str.contains("AW"))
]["gameId"].unique()
```

```python
for gameId in sorted(unique_game_id):
    print(gameId)
```
